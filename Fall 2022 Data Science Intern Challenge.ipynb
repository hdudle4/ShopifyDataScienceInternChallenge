```python
#Seeing Adjusted AOV Based on total orders divided by total items
total_revenue = df['order_amount'].sum()
total_items_ordered = df['total_items'].sum()

adjusted_AOV = total_revenue / total_items_ordered
adjusted_AOV
```

```python
#Seeing Adjusted AOV per shopID
adjusted_AOV_grouped = df.groupby('shop_id').order_amount.sum() / df.groupby('shop_id').total_items.sum()
adjusted_AOV_grouped.sort_values()
```

```python
#Seeing Adjusted AOV without shop 78
df_exclusive= adjusted_AOV_grouped.sort_values()[:-1]
df_exclusive.mean()
```

The AOV is now adjusted more appropriately per shop. The data is much less skewed because Shop 78's AOV is scaled much closer to the rest of the shops' AOV. Shop 78 is going to be an outlier. So we did the AOV without Shop 78, with the the AOV being 152.26. The AOV for Shop 78 is 25725

```python
#Seeing the Median of the Adjusted Order Value
adjusted_AOV_grouped.median()

```

Taking the median of the AOV is another way of seeing how accurate the Adjusted AOV is. This gauges is very close to the original AOV, so the median of the Adjusted AOV is an accurate way of telling if the Adjusted AOV is a good estimate.

B:

SELECT LastName, COUNT(Orders.EmployeeID) 
FROM Orders 
INNER JOIN Employees 
ON Orders.EmployeeID = Employees.EmployeeID 
GROUP BY Orders.EmployeeID 
ORDER BY COUNT(Orders.EmployeeID) DESC 
LIMIT 1;

Peacock --> 40 orders

SELECT ProductName, SUM(Quantity) as Total FROM Orders AS o
INNER JOIN OrderDetails AS d ON d.OrderID = o.OrderID
INNER JOIN Products AS p ON p.ProductID = d.ProductID
INNER JOIN Customers AS c ON c.customerID = o.CustomerID 
WHERE Country = "Germany"
GROUP BY p.ProductID
ORDER BY Total DESC
LIMIT 1;
